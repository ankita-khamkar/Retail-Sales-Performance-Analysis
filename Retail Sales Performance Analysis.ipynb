```python
#check cancelled orders

cancelled_orders = retail_df[
    retail_df["InvoiceNo"].astype(str).str.startswith("C")
    ]
print("Cancelled Orders:", cancelled_orders.shape[0])
```

```python
# Remove Negative Quantities
retail_df = retail_df[retail_df["Quantity"] > 0]
```

```python
# Check Zero or Negative Prices

retail_df[retail_df["UnitPrice"] <=0]
```

```python
# Remove zero or negative prices

retail_df = retail_df[retail_df["UnitPrice"] > 0]
```

```python
# Verify dataset shape

print("Dataset Shape after Cleaning:", retail_df.shape)
```

```python
# Create Total Sales Column

retail_df["TotalSales"] = retail_df["Quantity"] * retail_df["UnitPrice"]
```

```python
retail_df[["Quantity", "UnitPrice", "TotalSales"]].head()
```

```python
# Extract Year

retail_df["Year"] = retail_df["InvoiceDate"].dt.year
```

```python
# Extract Month Number

retail_df["Month"] = retail_df["InvoiceDate"].dt.month
```

```python
# Extract Month Name

retail_df["MonthName"] = retail_df["InvoiceDate"].dt.month_name()
```

```python
#Extract Day

retail_df["Day"] = retail_df["InvoiceDate"].dt.day
```

```python
# Extract Day Name

retail_df["DayName"] = retail_df["InvoiceDate"].dt.day_name()
```

```python
# Extract Hour

retail_df["Hour"] = retail_df["InvoiceDate"].dt.hour
```

```python
# Extract Quarter

retail_df["Quarter"] = retail_df["InvoiceDate"].dt.quarter
```

```python
# Extract Weekday Number

retail_df["WeekDay"] = retail_df["InvoiceDate"].dt.weekday
```

# Sales Trend Analysis

```python
# Monthly Revenue Trend

monthly_sales = retail_df.groupby("MonthName", observed=True)["TotalSales"].sum().reset_index()

plt.figure(figsize=(12,6))

plt.plot(
    monthly_sales["MonthName"],
    monthly_sales["TotalSales"],
    marker="o",
    linewidth=2
)

plt.title("Monthly Revenue Trend", fontsize=16)
plt.xlabel("Month")
plt.ylabel("Revenue (£)")
plt.xticks(rotation=45)
plt.grid(True)

plt.show()
```

```python
# Monthly Revenue

plt.figure(figsize=(12,6))

plt.bar(
    monthly_sales["MonthName"],
    monthly_sales["TotalSales"]
)

plt.title("Monthly Revenue")
plt.xlabel("Month")
plt.ylabel("Revenue (£)")
plt.xticks(rotation=45)

plt.show()
```

```python
# Quartely Revenue

quarter_sales = retail_df.groupby("Quarter")["TotalSales"].sum().reset_index()

plt.figure(figsize=(8,5))

plt.bar(
    quarter_sales["Quarter"].astype(str),
    quarter_sales["TotalSales"]
)

plt.title("Quarter-wise Revenue")
plt.xlabel("Quarter")
plt.ylabel("Revenue (£)")

plt.show()
```

# Country Analysis

```python
# Top 10 Countries by Revenue

country_sales = retail_df.groupby("Country")["TotalSales"].sum().nlargest(10).reset_index()

plt.figure(figsize=(12,6))

plt.barh(
    country_sales["Country"],
    country_sales["TotalSales"]
)

plt.title("Top 10 Countries by Revenue")
plt.xlabel("Revenue (£)")
plt.ylabel("Country")

plt.show()
```

# Customer Analysis

```python
# Top 10 Customers

top_customers = (
    retail_df.groupby("CustomerID")["TotalSales"]
    .sum()
    .nlargest(10)
    .reset_index()
)

plt.figure(figsize=(12,6))

plt.bar(
    top_customers["CustomerID"].astype(str),
    top_customers["TotalSales"]
)

plt.title("Top 10 Customers by Revenue")
plt.xlabel("Customer ID")
plt.ylabel("Revenue (£)")

plt.show()
```

# Product Analysis

```python
# Top 10 Products

top_products = (
    retail_df.groupby("Description")["TotalSales"]
    .sum()
    .nlargest(10)
    .reset_index()
)

plt.figure(figsize=(14,6))

plt.barh(
    top_products["Description"],
    top_products["TotalSales"]
)

plt.title("Top 10 Products by Revenue")
plt.xlabel("Revenue (£)")
plt.ylabel("Product")

plt.show()
```

```python
# Quantity Sold by Product

quantity_products = (
    retail_df.groupby("Description")["Quantity"]
    .sum()
    .nlargest(10)
    .reset_index()
)

plt.figure(figsize=(14,6))

plt.barh(
    quantity_products["Description"],
    quantity_products["Quantity"]
)

plt.title("Top 10 Products by Quantity Sold")
plt.xlabel("Quantity")
plt.ylabel("Product")

plt.show()
```

# Time Analysis

```python
# Hourly Sales

hourly_sales = retail_df.groupby("Hour")["TotalSales"].sum().reset_index()

plt.figure(figsize=(12,5))

plt.plot(
    hourly_sales["Hour"],
    hourly_sales["TotalSales"],
    marker="o"
)

plt.title("Revenue by Hour")
plt.xlabel("Hour")
plt.ylabel("Revenue (£)")
plt.grid(True)

plt.show()
```

```python
# Orders by Weekday

weekday_orders = retail_df.groupby("DayName", observed=True)["InvoiceNo"].nunique().reset_index()

plt.figure(figsize=(10,5))

plt.bar(
    weekday_orders["DayName"],
    weekday_orders["InvoiceNo"]
)

plt.title("Orders by Weekday")
plt.xlabel("Day")
plt.ylabel("Orders")

plt.show()
```
